## 1. Magnetosheath

```python
filename = 'magnetosheath5s.DAT'
df = convertDframe(filename)
df.head()
```

```python
df.describe()
```

```python
# Load other libraries
import matplotlib.pyplot as plt
```

```python
# magnetic components
mag = [df['Br'],df['Bt'],df['Bn'],df['Bmag']]

# Plot numerical data
a = 2 # number of rows
b = 2 # number of columns
c = 1 # initialize plot counter

fig = plt.figure(figsize = (15,5))

for i in mag:
    plt.subplot(a,b,c)
    plt.plot(df['A'], i, color = 'black')
    plt.title('Magnetosheath - ' + i.name)
    plt.xlabel('December 2000', fontsize = 12)
    plt.ylabel(i.name, fontsize = 12)
    plt.yticks(fontsize = 12)
    plt.tight_layout()
    c = c + 1

plt.show()
```

```python
# Está sendo analisando um total de 12960 pontos. O desvio padrão das componentes de Br e Bmag são os que 
#se encontram mais próximo.
```
